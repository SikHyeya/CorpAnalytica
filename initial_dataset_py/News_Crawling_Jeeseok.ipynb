```python
# https://www.etoday.co.kr

import requests
from bs4 import BeautifulSoup

def crawl_news_article_without_relation(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # <div> 태그 중 클래스가 'relation_newslist'인 요소를 찾아서 제외
        relation_div = soup.find('div', class_='relation_newslist')
        if relation_div:
            relation_div.extract()
        
        # 원문을 포함한 HTML 요소의 선택자를 확인하여 적절히 수정하세요
        article_element = soup.select_one('.articleView')  # 예시 선택자
        if article_element:
            article_text = article_element.get_text()
            return article_text
    return None

news_url = 'https://www.etoday.co.kr/news/view/2274768'
article_text = crawl_news_article_without_relation(news_url)

if article_text:
    print(article_text)
else:
    print("뉴스 원문을 가져올 수 없습니다.")
```

```python
# www.newspim.com 
import requests
from bs4 import BeautifulSoup

def crawl_news_contents(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        news_contents_element = soup.find('div', id='news-contents')
        if news_contents_element:
            news_contents = news_contents_element.get_text()
            return news_contents
    return None

news_url = 'https://www.newspim.com/news/view/20230811000655'
news_contents = crawl_news_contents(news_url)

if news_contents:
    print(news_contents)
else:
    print("뉴스 내용을 가져올 수 없습니다.")
```
